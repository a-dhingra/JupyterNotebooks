```python
import pandas as pd
import timeit
import sys, os
import numpy as np
```

```python
file = r'/content/drive/MyDrive/Data/H1B_Companies.csv'
```

```python
df = pd.read_csv(file)
df.head()
```

```python
df1 = df.iloc[1:].copy()
df1.head()
```

```python
list(df1)
```

```python
df1 = df1[[' Company_Name',' No_of_H1B_Filings',' Average_Salary']]
df1.head()
```

```python
df1 = df1.rename(columns={' Company_Name':'Company_Name'
                          ,' No_of_H1B_Filings':'No_of_H1B_Filings'
                          ,' Average_Salary':'Average_Salary'
                          })
df1.head()
```

```python
list(df1)
```

```python
df1['Salary'] = df1['Average_Salary'].apply(lambda x: x[1:]).apply(lambda x: x.replace(',',''))
df1['No_of_H1B_Filings'] = df1['No_of_H1B_Filings'].apply(lambda x: x.replace(',',''))
df1.head()
```

```python
df1.info()
```

```python
df1['num'] = df1['No_of_H1B_Filings'].astype(float) * df1['Salary'].astype(float)
df1.head()
```

```python
# sorted_df = df_marks.sort_values(by='algebra')
df2 = df1.sort_values(by='num',ascending=False)
df2.head(10)
```

```python
df2.loc[df2['Salary'].astype(float) > 100000].iloc[351:1500,:].sort_values(by='Salary', ascending=False).iloc[:10,:]
```
