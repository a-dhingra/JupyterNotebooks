## Next Steps: 
- Use the 'Cpu Name' column to identify Cpu names as Intel or AMD or other.
  - Intel (i3,i5,i7) and AMD models needs to be identified. Assign a numeric value to each.
  - Also, create a new column capturing the Ghz speed of the Cpus
- Create a column like 'Has_SSD' and compute the same using the 'Memory' column. 'Flash Storage' means 'SSD' too.
- Create a column like 'SSD-Storage' to capture the amount of SSD
- Try to get the HDD value and use it (if possible)
- Assign a numeric value to each 'OpSys'
- Use 'Company' column as well to understand the correlation between brand and price.

```python
# Function to compute if a model has SSD/Flash storage or not
def has_SSD(x):
  if 'SSD' in x:
    return 1
  elif 'Flash' in x:
    return 1
  else:
    return 0


# Function to give weight to the different Operating Systems 
def OpSys_weight(x):
  if 'Windows' in x:
    return 1
  elif 'mac' in x or 'Mac' in x:
    return 2
  elif 'Chrome' in x or 'Android' in x:
    return 3
  elif 'Linux' in x:
    return 4
  else:
    return 0

  

```

```python
  # Funtion to compute the CPU weights
  def intel_or_AMD(x):
    if 'Intel' in x:
      return 1
    elif 'AMD' in x:
      return 0
    else:
      return 2 
```

```python
# Creating a Dictionary to assign a numeric value to each 'Company' and 'Type Name'

company_list = list(df['Company'].unique())
typeName_list = list(df['TypeName'].unique())
company_list
```

```python
typeName_list
```

```python
company_index = [x for x in range(len(company_list))]
company_index
```

```python
typeName_index = [x for x in range(len(typeName_list))]
typeName_index
```

```python
company_dict = dict(zip(company_list, company_index))
company_dict
```

```python
typeName_dict = dict(zip(typeName_list, typeName_index))
typeName_dict
```

```python
df['has_SSD'] = df['Memory'].apply(lambda x: has_SSD(x))
df['OpSys_weight'] = df['OpSys'].apply(lambda x: OpSys_weight(x))
df['Intel_or_AMD'] = df['Cpu'].apply(lambda x: intel_or_AMD(x))
df['Company_weight'] = df['Company'].apply(lambda x: company_dict.get(x))
df['TypeName_weight'] = df['TypeName'].apply(lambda x: typeName_dict.get(x))
df.head()
```

```python
# Creating a new DataFrame with only the required columns for building the ML Model

df_ml = df[['Price','Weight','TouchScreen','Ips','ppi','has_SSD','OpSys_weight','Intel_or_AMD','Company_weight','TypeName_weight']]
df_ml.info()
```

```python
df_ml.isnull().sum()
```

### Building ML Model (Linear Regression)

```python
sns.pairplot(df_ml)
```

```python
x = df_ml[['Weight','TouchScreen','Ips','ppi','has_SSD','OpSys_weight','Intel_or_AMD','Company_weight','TypeName_weight']]
y = df_ml[['Price']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
# Importing Linear Regrsssion Model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.coef_
```

```python
print(model.intercept_)
```

```python
list(x.columns)
```

```python
pd.DataFrame(list(x.columns), list(model.coef_))
```

### Predictions

```python
predictions = model.predict(x_test)
```

```python
plt.figure(figsize=[8,6])
plt.scatter(y_test, predictions, alpha=0.65, linewidths=1)
```

```python
plt.figure(figsize=[10,8])
sns.histplot(y_test-predictions, fill=True)
```
